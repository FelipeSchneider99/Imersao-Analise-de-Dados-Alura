```python
senioridade = {
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
}

df['senioridade'] = df['senioridade'].replace(senioridade)
df['senioridade'].value_counts()
```

```python
tamanho_empresa = {
    'M': 'Mdio',
    'L': 'Grande',
    'S': 'Pequeno'
}

df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa)
df['tamanho_empresa'].value_counts()
```

```python
remoto = {
    0: 'Presencial',
    50: 'Hbrido',
    100: 'Remoto'
}

df['remoto'] = df['remoto'].replace(remoto)
df['remoto'].value_counts()
```

```python
import numpy as np

# criao de um dataframe de teste para usar de exemplo

df_salarios = pd.DataFrame({
    'nome' : ['Felipe', 'Olivia', 'Shoyo', 'Tofu', 'Gatao'],
    'salario' : [np.nan, 5000, 8000, np.nan, 98654]
})

# calcula media e substitui os nulos pela media e arredonda os valores

df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

# calcula a mediana e substitui os nulos pela mediana

df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    "Dia" : ['Segunda', 'Tera', 'Quarta', 'Quinta', 'Sexta'],
    "Temperatura" : [30, np.nan, np.nan, 28, 27]
})

#prenche os valores nulos com o valor anterior / forward fill

df_temperaturas['Preenchido_ffill'] = df_temperaturas["Temperatura"].ffill()
df_temperaturas
```

```python
#prenche os valores nulos com o valor seguinte / backward fill

df_temperaturas['Preenchido_bfill'] = df_temperaturas["Temperatura"].bfill()
df_temperaturas
```

```python
df_cidades = pd.DataFrame({
    'nome' : ['Felipe', 'Olivia', 'Shoyo', 'Tofu', 'Gatao'],
    'cidade' : ['Noia City', np.nan, 'Karumai', 'Osaka', np.nan]
})

df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('No informado')
df_cidades
```

Bibliotecas para construo de grficos Python - MATPLOTLIB / SEABORN

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
sns.barplot(data=df_limpo, x='senioridade', y='usd')
```

```python
# Definir tamanho do grafico
plt.figure(figsize= (8, 5))
sns.barplot(data=df_limpo, x='senioridade', y='usd')

# Titulo do Grafico PLT
plt.title('Salrio mdio por nvel de Senioridade')

# Nomear eixo x
plt.xlabel('Nivel de senioridade')

# Nomear eixo y
plt.ylabel('Salario mdio anual (USD)')

# Mostrar o grfico
plt.show()
```

```python
# Sort do maior para o menor
# Ascending False = Decrescente // Ascending True = Crescente
df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending=True)

# Declarando varivel para manter os valores do sort
ordem = df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending=True).index
ordem
```

```python
# Definir tamanho do grafico
plt.figure(figsize= (8, 5))
sns.barplot(data=df_limpo, x='senioridade', y='usd', order = ordem)

# Titulo do Grafico PLT
plt.title('Salrio mdio por nvel de Senioridade')

# Nomear eixo x
plt.xlabel('Nivel de senioridade')

# Nomear eixo y
plt.ylabel('Salario mdio anual (USD)')

# Mostrar o grfico
plt.show()
```

```python
plt.figure(figsize= (8, 5))

# Histograma

# KDE = Linhazinha
# Bins =  Largura das barras / Quanto maior o valor, mais granulado
sns.histplot(df_limpo['usd'], bins= 50, kde = True)
plt.title('Distribuio de salarios anuais')
plt.xlabel('Salario em USD')
plt.ylabel('Frequencia')
plt.show()
```

```python
plt.figure(figsize= (8, 5))

# Boxplot

sns.boxplot(x = df_limpo['usd'])
plt.title('Boxplot Salario')
plt.xlabel('Salario em USD')

plt.show()
```

```python
ordem_senioridade = ['Junior', 'Pleno', 'Senior', 'Executivo']
plt.figure(figsize= (8,5))
sns.boxplot(x = 'senioridade', y = 'usd', data = df_limpo, order = ordem_senioridade)
plt.title('Boxplot da distribuio por senioridade')
plt.xlabel('Senioridade')
plt.ylabel('Salario em USD')
plt.show()
```

```python
ordem_senioridade = ['Junior', 'Pleno', 'Senior', 'Executivo']
plt.figure(figsize= (8,5))
sns.boxplot(x = 'senioridade', y = 'usd', data = df_limpo, order = ordem_senioridade, palette= 'Set2', hue= 'senioridade')
plt.title('Boxplot da distribuio por senioridade')
plt.xlabel('Senioridade')
plt.ylabel('Salario em USD')
plt.show()
```

```python
df_data_scientist = df_limpo[df_limpo['cargo'] == 'Data Scientist']
salario = df_data_scientist.groupby('residencia')['usd'].mean().reset_index()


fig = px.bar(salario,
             x= 'residencia',
             y= 'usd',
             title= 'Salrio Data Scientist no Pas de Residncia',
             labels={'residencia': 'Pas de Residncia', 'usd': 'Salrio do Cargo'})
fig.update_layout(xaxis={'categoryorder': 'total ascending'})
fig.show()
```

```python
import country_converter as coco

# Cria uma lista das siglas dos pases presentes na coluna 'residencia' do DataFrame salario
country_codes = salario['residencia'].tolist()

# Converte a lista de siglas de pases para os nomes oficiais completos
country_full_names = coco.convert(names=country_codes, to='name_official')

# Cria um dicionrio onde as chaves so as siglas dos pases e os valores so os nomes completos
country_mapping = dict(zip(country_codes, country_full_names))

# Adiciona uma nova coluna com o nome completo dos pases para o hover
salario['nome_pais'] = salario['residencia'].map(country_mapping)

fig = px.bar(salario,
             x= 'residencia',
             y= 'usd',
             title= 'Salrio Data Scientist',
             labels={'residencia': 'Pas de Residncia do Funcionrio', 'usd': 'Salrio Anual'},
             color= 'residencia',
             # Mostra o nome do pas no hover
             hover_name='nome_pais'
            )

# Ordena o grfico de forma ascendente
fig.update_layout(xaxis={'categoryorder':'total ascending'})

fig.show()
```
